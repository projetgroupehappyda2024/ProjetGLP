```python
# Dans la df happy on a choisi d'ajouter un indicateur de la liberté avec la df freedom
happy = pd.read_excel('DataForTable2.1.xls') 
freedom = pd.read_excel('2023-Human-Freedom-Index-Data.xlsx')
happy.head()
```

```python
# Nous renommons les colonnes pour avoir le meme nom dans les 2 tableaux
freedom=freedom.rename(columns={'Unnamed: 2':'Country name'})
freedom=freedom.rename(columns={'Unnamed: 0':'year'})
# Fusion pour integrer dans happy la colonne human freedom du DF freedom
happy = pd.merge(happy, freedom[['year','Country name','PERSONAL FREEDOM','ECONOMIC FREEDOM']], on=['year', 'Country name'], how='left')
happy = happy.rename(columns={'PERSONAL FREEDOM':'Personal Freedom'})
happy = happy.rename(columns={'ECONOMIC FREEDOM':'Economic Freedom'})
happy.head()
```

```python
# Code pour repérage et ajout des sous-régions par pays
from countryinfo import CountryInfo

# Fonction pour obtenir la sous-région d'un pays
def get_country_info(country_name):
    try:
        # Création d'une instance de CountryInfo pour le pays donné
        country = CountryInfo(country_name)
        info = country.info()

        # Obtention de la sous-région
        regions = info.get('subregion')

        return regions
    except KeyError:
        return None

# Application de la fonction pour obtenir la sous-région pour chaque pays
happy['Regional indicator'] = happy['Country name'].apply(get_country_info)
```

```python
# Répartition des continents en fonction de la sous-région
continent_repartition = {
    'Africa': ['Northern Africa', 'Middle Africa', 'Western Africa', 'Southern Africa', 'Eastern Africa'],
    'North America': [ 'Northern America','Caribbean'],
    'South America': ['South America', 'Central America'],
    'Asia': ['Southern Asia', 'Western Asia', 'South-Eastern Asia', 'Eastern Asia', 'Central Asia'],
    'Oceania': ['Australia and New Zealand'],
    'Europe': ['Southern Europe', 'Western Europe', 'Eastern Europe', 'Northern Europe']
}

# Fonction pour compléter la colonne "continent"
def attribution_continent(sous_region):
    for continent, regions in continent_repartition.items():
        if sous_region in regions:
            return continent
    return 'NaN'  # Si pas de continent conrrespondant

# Application de la fonction d'attribution du continent en fonction des sous-régions
happy['Continent'] = happy['Regional indicator'].apply(attribution_continent)

# Affichage du DataFrame avec la colonne 'Continent' 
happy.head(30)
```

```python
# Le nom des pays n'est pas formaté pour être reconnu par CountryInfo.
# Pour l'utilisation de la liste de CountryInfo (https://gist.github.com/mjrulesamrat/0c1f7de951d3c508fb3a20b4b0b33a98)
# nous avons modifié les noms pays avec des nan

# Dictionnaire de correspondance des noms de pays
modificationNomPays_dico = {
    'Congo (Brazzaville)': 'Republic of the Congo',
    'Congo (Kinshasa)': 'Democratic Republic of the Congo',
    'Czechia': 'Czech Republic',
    'Eswatini': 'Eswatini (Swaziland)',
    'Gambia': 'The Gambia',
    'Hong Kong S.A.R. of China': 'Hong Kong',
    'Kosovo': 'Kosovo',
    'Montenegro': 'Montenegro',
    'Myanmar (Burma)': 'Myanmar',
    'North Macedonia': 'Macedonia',
    'Somaliland region': 'Somaliland',
    'State of Palestine': 'Palestine',
    'Taiwan Province of China': 'Taiwan',
    'Türkiye': 'Turkey'
}


# Remplacement uniquement des pays présents dans le dictionnaire
happy['Country name'] = happy['Country name'].replace(modificationNomPays_dico)
```

```python
# Application de la fonction pour obtenir la sous-région pour chaque pays
happy['Regional indicator'] = happy['Country name'].apply(get_country_info)
```

```python
# Affichage des pays sans region
pays_sans_region = happy[happy['Regional indicator'].isna()]['Country name'].unique()
print(pays_sans_region)
```

```python
# Dictionnaire des sous-régions à affecter manuellement

manuel_sregions = {
    'Eswatini (Swaziland)': 'Southern Africa',
    'Kosovo': 'Southern Europe',
    'Montenegro': 'Southern Europe',
    'Myanmar': 'South-Eastern Asia',
    'Macedonia': 'Southern Europe',
    'Somaliland': 'Eastern Africa',
    'Palestine': 'Western Asia'
}

# Affectation des sous-régions manuellement pour les pays non reconnus par CountryInfo malgré la mise en conformité du nom du pays
happy['Regional indicator'] = happy.apply(lambda row: manuel_sregions.get(row['Country name'], row['Regional indicator']), axis=1)
```
