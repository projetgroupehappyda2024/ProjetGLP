```python
# Dans la df happy on a choisi d'ajouter un indicateur de la libert√© avec la df freedom
happy = pd.read_excel('DataForTable2.1.xls') 
freedom = pd.read_excel('2023-Human-Freedom-Index-Data.xlsx')
happy.head()
```

```python
# Nous renommons les colonnes pour avoir le meme nom dans les 2 tableaux
freedom=freedom.rename(columns={'Unnamed: 2':'Country name'})
freedom=freedom.rename(columns={'Unnamed: 0':'year'})
# Fusion pour integrer dans happy la colonne human freedom du DF freedom
happy = pd.merge(happy, freedom[['year','Country name','PERSONAL FREEDOM','ECONOMIC FREEDOM']], on=['year', 'Country name'], how='left')
happy = happy.rename(columns={'PERSONAL FREEDOM':'Personal Freedom'})
happy = happy.rename(columns={'ECONOMIC FREEDOM':'Economic Freedom'})
happy.head()
```

```python
# Code pour rep√©rage et ajout des sous-r√©gions par pays
from countryinfo import CountryInfo

# Fonction pour obtenir la sous-r√©gion d'un pays
def get_country_info(country_name):
    try:
        # Cr√©ation d'une instance de CountryInfo pour le pays donn√©
        country = CountryInfo(country_name)
        info = country.info()

        # Obtention de la sous-r√©gion
        regions = info.get('subregion')

        return regions
    except KeyError:
        return None

# Application de la fonction pour obtenir la sous-r√©gion pour chaque pays
happy['Regional indicator'] = happy['Country name'].apply(get_country_info)
```

```python
# R√©partition des continents en fonction de la sous-r√©gion
continent_repartition = {
    'Africa': ['Northern Africa', 'Middle Africa', 'Western Africa', 'Southern Africa', 'Eastern Africa'],
    'North America': [ 'Northern America','Caribbean'],
    'South America': ['South America', 'Central America'],
    'Asia': ['Southern Asia', 'Western Asia', 'South-Eastern Asia', 'Eastern Asia', 'Central Asia'],
    'Oceania': ['Australia and New Zealand'],
    'Europe': ['Southern Europe', 'Western Europe', 'Eastern Europe', 'Northern Europe']
}

# Fonction pour compl√©ter la colonne "continent"
def attribution_continent(sous_region):
    for continent, regions in continent_repartition.items():
        if sous_region in regions:
            return continent
    return 'NaN'  # Si pas de continent conrrespondant

# Application de la fonction d'attribution du continent en fonction des sous-r√©gions
happy['Continent'] = happy['Regional indicator'].apply(attribution_continent)

# Affichage du DataFrame avec la colonne 'Continent' 
happy.head(30)
```

```python
# Dictionnaire des sous-r√©gions √† affecter manuellement

manuel_sregions = {
    'Eswatini (Swaziland)': 'Southern Africa',
    'Kosovo': 'Southern Europe',
    'Montenegro': 'Southern Europe',
    'Myanmar': 'South-Eastern Asia',
    'Macedonia': 'Southern Europe',
    'Somaliland': 'Eastern Africa',
    'Palestine': 'Western Asia'
}

# Affectation des sous-r√©gions manuellement pour les pays non reconnus par CountryInfo malgr√© la mise en conformit√© du nom du pays
happy['Regional indicator'] = happy.apply(lambda row: manuel_sregions.get(row['Country name'], row['Regional indicator']), axis=1)
```
